```python
# Ejercicio 1

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 10000)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, color="green", label="sen(x)")
plt.plot(x, y_cos, color="blue", label="cos(x)")

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Grafica de seno y coseno")
plt.legend()
plt.show()
```

```python
#Ejercicio 2

# Define the quadratic function
def quadratic_function(x):
    return x**2

# Generate data points
x = np.linspace(0, 10, 10000)
y = quadratic_function(x)

# Plot the function
plt.plot(x, y)

# Set the limits of the axes
plt.xlim(0, 10)
plt.ylim(-10, 100)

# Add labels and title
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Funci√≥n cuadr√°tica")

# Show the plot
plt.show()
```

```python
 # Ejercicio 3

# Define the trigonometric function (e.g., tangent)
def trigonometric_function(x):
    return np.tan(x)

# Generate data points
x = np.linspace(-np.pi/2, np.pi/2, 10000)
y = trigonometric_function(x)

# Plot the function
plt.plot(x, y)

# Set the limits of the axes
plt.xlim(-np.pi/2, np.pi/2)
plt.ylim(-5, 5)

# Add labels and title
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Gr√°fica de la funci√≥n tangente")

# Add an annotation at a relevant point on the curve
plt.annotate("Punto importante", xy=(np.pi/4, np.tan(np.pi/4)), xytext=(np.pi/4, 4),
            arrowprops=dict(facecolor='yellow', shrink=0.05))

# Show the plot
plt.show()
```

```python
# Ejercicio 4

import pandas as pd

np.random.seed(4)

meses = range(1,13)
ventas_tecnologia = np.random.randint(130, 450, 12)
ventas_hogar = np.random.randint(200, 350, 12)
ventas_herramientas = np.random.randint(200, 500, 12)

fig = plt.figure(figsize = (12,8))


df_ventas = pd.DataFrame({"Meses": meses,
                         "Ventas tecnolog√≠a": ventas_tecnologia,
                         "Ventas hogar": ventas_hogar,
                         "Ventas herramientas": ventas_herramientas})

ax1 = plt.subplot2grid((2,2), (0,0))
ax2 = plt.subplot2grid((2,2), (0,1))
ax3 = plt.subplot2grid((2,2), (1,0), colspan=2)


ax1.plot(df_ventas['Meses'], df_ventas['Ventas tecnolog√≠a'], marker='o', color="orange")
ax1.grid(True)
ax1.set_xlabel("Meses")
ax1.set_ylabel("Ventas")
ax1.set_title('Ventas tecnolog√≠a 2023')


ventas_agosto = df_ventas.loc[df_ventas['Meses'] == 8]
ventas_agosto.set_index('Meses', inplace = True)
ax2.bar(ventas_agosto.columns, ventas_agosto.values[0], color = ['#C70039', '#FFC300','#4FFF33'], width= 0.3)
ax2.set_ylabel("Ventas")
ax2.set_title('Ventas mes de agosto 2023')

df_ventas.set_index('Meses', inplace = True)

ax3.pie(df_ventas.sum(),labels=df_ventas.columns, explode=(0, 0, 0.1),  autopct='%1.1f%%',
        shadow=True, startangle= 90)
ax3.set_title('Ventas 2023 por categoria')


plt.tight_layout()

plt.show()
```

```python
# Ejercicio 5

from scipy.ndimage import gaussian_filter

np.random.seed(0)
matrix = np.random.rand(4, 4)
gaussian_matrix = gaussian_filter(matrix, sigma=1)

plt.figure(figsize=(8, 6))
plt.imshow(gaussian_matrix, interpolation='gaussian', cmap='plasma')
plt.colorbar()
plt.grid(True)
plt.title('Mapa de Calor con Interpolaci√≥n Gaussiana')
plt.show()
```
